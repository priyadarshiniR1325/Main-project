
# Employee Engagement & Workplace Satisfaction Analysis



# Objective

1. To analyze employee survey data to understand overall levels of job satisfaction, work-life balance, and workplace experience.
2. To identify key drivers of employee engagement and dissatisfaction across departments, job levels, and demographic groups.
3. To examine the relationship between work environment, workload, compensation-related factors, and employee satisfaction.

# Types of Analysis

1. Descriptive Analysis (What is happening?)
•	Total employees, average job satisfaction, experience, and overtime rates
•	Workforce composition by gender, age group, department, marital status, education level, and job level
•	Distribution of experience, training hours, and reporting structures
•	Average engagement, work environment, and work-life balance scores across departments and roles.
             
2. Diagnostic Analysis (Why is it happening?)
•	Comparison of job satisfaction across departments, job levels, experience groups, and work environment scores
•	Analysis of overtime impact on satisfaction and well-being
•	Exploration of training hours vs satisfaction and experience
•	Identification of departments or roles with high workloads and low engagement

 3. Predictive Analysis (What is likely to happen?)
•	Predicts which employee groups are more likely to experience low satisfaction or burnout (e.g., high overtime + low WLB)
•	Identifies experience or training patterns associated with higher engagement
•	Forecasts workforce risks by department or job level based on historical trends

4. Prescriptive Analysis (What should be done?)
•	Targeted training investment for roles with low satisfaction but high growth potential
•	Workload redistribution in departments with high overtime and low well-being
•	Career path redesign where experience does not translate into job progression
•	Engagement interventions for low-performing teams

## Chart 1- Job Satisfaction Distribution

```python
import matplotlib.pyplot as plt
plt.hist(df["JobSatisfaction"], bins=5)
plt.title("Distribution of Job Satisfaction")
plt.xlabel("Job Satisfaction Score")
plt.ylabel("Employee Count")
plt.show()
```

# Interpretation
Most employees rate job satisfaction between 3 and 5, indicating generally positive engagement, but a noticeable group reports lower satisfaction, signaling potential improvement areas.

# Chart 2 Department Count

```python
df["Dept"].value_counts().plot(kind="bar")
plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

## Interpretation
Some departments have significantly larger employee populations, meaning engagement initiatives in these areas will impact the largest portion of the workforce.


# Chart  Work-Life Balance vs Job Satisfaction

```python
plt.scatter(df["WLB"], df["JobSatisfaction"])
plt.xlabel("Work-Life Balance")
plt.ylabel("Job Satisfaction")
plt.title("WLB vs Job Satisfaction")
plt.show()
```

## Interpretation
There is a strong positive relationship — as work-life balance improves, job satisfaction also increases substantially.

# Chart 4 Job Satisfaction by Department

```python
df.groupby("Dept")["JobSatisfaction"].mean().plot(kind="bar")
plt.title("Average Job Satisfaction by Department")
plt.ylabel("Avg Job Satisfaction")
plt.show()
```

## Interpretation
Some departments consistently report lower satisfaction levels, making them priority targets for HR intervention.



# Chart 5 Job Satisfaction by Gender & Department

```python
pd.pivot_table(df, values="JobSatisfaction", index="Dept", columns="Gender").plot(kind="bar")
plt.title("Job Satisfaction by Dept and Gender")
plt.ylabel("Avg Satisfaction")
plt.show()
```

# Interpretation
Gender-based differences are minimal overall, but certain departments show small gaps, suggesting targeted inclusion initiatives could improve engagement.
